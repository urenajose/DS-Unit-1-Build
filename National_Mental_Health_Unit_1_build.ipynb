##Exploration of df features

```
df_main_services = df.copy() #so create a df without ['FACILITYTYPE', 'FOCUS', 'OWNERSHP', 'PUBLICAGENCY', 'LANGPROV','SMOKINGPOLICY'] multi level questions only main features.
```

```
df_main_services = df.drop(["CASEID","LST",'FACILITYTYPE', 'FOCUS', 'OWNERSHP', 'PUBLICAGENCY', 'LANGPROV',
       'SMOKINGPOLICY'], axis=1)

df_main_services = df_main_services[(df_main_services == 1)]
df_main_services.insert(0,"LST",df["LST"])
```

```
dfMH_group_states = df_main_services.groupby("LST").sum() #grouping MH services by state.
```

```
dfMH_group_states = dfMH_group_states.reset_index() #save the grouping into a variable and add LST as a column, create a index.
```

```
dfMH_group_states #Sum of services provided (Questionare question = 1), by feature
```

###PCA  method applied to Sum of services provided (Questionare question = 1), by feature dfMH_group_states.

```
import matplotlib.pyplot as plt
# import the PCA module from sklearn
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```
feature_cols_dfmhgs = dfMH_group_states.columns[1:] 
feature_cols_dfmhgs #column labels from dfMH_group_states DataFrame
```

```
x = dfMH_group_states.loc[:,feature_cols_dfmhgs].values
```

```
x = StandardScaler().fit_transform(x)
```

```
normal_dfMH_group_states = pd.DataFrame(x,columns=feature_cols_dfmhgs)
```

```
#print(normal_dfMH_group_states.head()) #to chek my normalized df
```

```
pca_dfMHGS = PCA(n_components=2)
```

```
pca_dfMHGS_data = pca_dfMHGS.fit_transform(x) #PCA DATA
```

```
dfMHGS = pd.DataFrame(data = pca_dfMHGS_data, columns=['comp_1','comp_2'])
```

```
sum(pca_dfMHGS.explained_variance_ratio_) #how much variance did we explained
```

```
1-sum(pca_dfMHGS.explained_variance_ratio_) 
```

```
dfMHGS['LST'] = dfMH_group_states['LST'] #drop the state column
dfMHGS = dfMHGS.drop(52) #drop "zz" column 
```

```
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(1, figsize=(20,10))
ax.scatter(dfMHGS['comp_1'], dfMHGS['comp_2'], s=1000, alpha=.2,lw=2)
fig.text(.13,.9,f'''{mhf} Group by: US states, Distric of Columbia and Puerto Rico
and the sum of survices they offer''',fontsize=20,fontweight="bold",color='#36A3F0')
ax.axes.xaxis.set_ticklabels([])
ax.axes.yaxis.set_ticklabels([])
for i, txt in enumerate(dfMHGS['LST']):
    ax.annotate(txt, (dfMHGS['comp_1'][i], dfMHGS['comp_2'][i]),fontsize=10)
plt.show()
```
